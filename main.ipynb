```python
@app.callback(
    Output('wc1', 'src'),
    [Input('submit-val', 'n_clicks')],
    state=[State('input1', 'value')
             ])
def update_output(n_clicks,value):
    df = scrape_by_keyword(value)
    source = create_wordcloud(df, mask,100)
    return source
```
